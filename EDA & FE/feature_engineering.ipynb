```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
pd.pandas.set_option('display.max_columns',None)
pd.pandas.set_option('display.max_rows',None)
```

```python
#applying on train data
ds=pd.read_csv('test.csv')
ds.head()
```

```python
ds.drop(['Id'],axis=1,inplace=True)
```

## handling missing values

```python
#categorical feature with nan values
nan_feature=[feature for feature in ds.columns if ds[feature].isnull().sum()>1 and ds[feature].dtypes=='O']
for feature in nan_feature:
    print(f'{feature} has {ds[feature].isnull().sum()} missing values, which is{np.round(ds[feature].isnull().mean(),2)} ')
```

```python
#fill missing categorical values with 'missing'
ds[nan_feature]=ds[nan_feature].fillna('missing')
ds[nan_feature].isnull().sum()
```

```python
ds.head()
```

```python
#numerical feature with missing values
num_with_nan=[feature for feature in ds.columns if ds[feature].isnull().sum()>1 and ds[feature].dtype!='O']
for feature in num_with_nan:
    print(f'{feature} has {ds[feature].isnull().sum()} missing values, which is {np.round(ds[feature].isnull().mean(),2)}')
```

```python
#fill missing values with median
for feature in num_with_nan:
    median=ds[feature].median()
    ds[feature].fillna(median,inplace=True)
ds[num_with_nan].isnull().sum()
```

```python
ds.head()
```

```python
ds['YearBuilt']=ds['YrSold']-ds['YearBuilt']
```

```python
ds['YearRemodAdd']=ds['YrSold']-ds['YearRemodAdd']
```

```python
ds['GarageYrBlt']=ds['YrSold']-ds['GarageYrBlt']
```

```python
ds[['YearBuilt','YearRemodAdd','GarageYrBlt']].head()
```

## Numerical Variables

```python
#apply log for normal distribution
feature_for_log=['LotFrontage','LotArea','1stFlrSF','GrLivArea']
for feature in feature_for_log:
    ds[feature]=np.log(ds[feature])
```
