```python
#importing libraris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
pd.pandas.set_option('display.max_columns',None)
```

```python
#loading dataset
ds=pd.read_csv('train.csv')
ds.shape
```

```python
ds.head()
```

```python
ds.tail()
```

## checking for missing values

```python
features_with_na=[features for features in ds.columns if ds[features].isnull().sum()>1]
features_with_na
```

```python
for features in features_with_na:
    print(features,ds[features].isnull().sum())
```

```python
#how much price varies with missing values
for feature in features_with_na:
    dst=ds.copy()
    dst[feature]=np.where(dst[feature].isnull(),1,0)
    dst.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

```python
ds.drop(['Id'],axis=1,inplace=True)
```

##  handling features with numerical values

```python
num_features=[features for features in ds.columns if ds[features].dtypes!='O']
```

```python
len(num_features)
```

```python
ds[num_features]
```

```python
#features with year value
yr_features=[feature for feature in num_features if 'Yr' in feature or 'Year' in feature]
yr_features
```

plot graph with all year feature

```python
for year_column in yr_features:
    ds.groupby(year_column)['SalePrice'].median().plot()
    plt.xlabel(year_column)
    plt.ylabel('Meadian of House Price')
    plt.title('House Price with {year_column}')
    plt.show()
```

## Discreate Variable

```python
discrete_features=[feature for feature in num_features if len(ds[feature].unique())<25 and feature not in yr_features]
discrete_features,len(discrete_features)
```

```python
ds[discrete_features].head()
```

```python
for feature in discrete_features:
    ds2=ds.copy()
    ds2.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

## Continuous Variable

```python
continuous_features=[feature for feature in num_features if feature not in discrete_features+yr_features]
continuous_features,len(continuous_features)
```

```python
for feature in continuous_features:
    ds2=ds.copy()
    ds2[feature].hist()
    plt.xlabel(feature)
    plt.ylabel('count')
    plt.title(feature)
    plt.show()
#we need to find out distribution of continuous values so we are using histogram
```

```python
#we use logarithmic transformation
for feature in continuous_features:
    ds2=ds.copy()
    if 0 in ds2[feature].unique():
        pass
    else:
        ds2[feature]=np.log(ds2[feature])
        ds2['SalePrice']=np.log(ds2['SalePrice'])
        plt.scatter(ds2[feature],ds2['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
        
```

```python
#checking outliers in continuous_feature
for feature in continuous_features:
    ds2=ds.copy()
    if 0 in ds2[feature].unique():
        pass
    else:
        ds2[feature]=np.log(ds2[feature])
        ds2.boxplot(column=feature)
        plt.ylabel('value')
        plt.title(feature)
        plt.show()
```
